```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import img_as_uint, img_as_float, img_as_int
from skimage.io import imread
from skimage.color import rgb2gray, gray2rgb
from scipy.fftpack import fft2, fftshift, ifft2, ifftshift
from skimage.filters import threshold_otsu
from skimage.measure import label, regionprops
from skimage.color import label2rgb
from skimage.draw import circle
from matplotlib.patches import Rectangle
import copy

%matplotlib inline
```

```python
def show(collection):
    fig, ax = plt.subplots(nrows=1, ncols=len(collection), figsize=(10, 10))
    for i in range(len(collection)):
        ax[i].imshow(collection[i])
        ax[i].axis('off')
```

```python
def get_regions(filename):
    klad0 = imread("images/Klad02.jpg")
    gray_klad0 = rgb2gray(klad0)
    binarized = np.where(gray_klad0 > 0.05, 1, 0)
    show([gray2rgb(klad0), gray2rgb(binarized), label2rgb(label(binarized))])
    regions = regionprops(label(binarized),  intensity_image=gray_klad0)
```

```python
def get_red_arrow(regions):
    for region in regions:
        red_arrow = False
        for i,j in region.coords:
            if klad0[i, j][0] > klad0[i, j][2] + 200 and klad0[i, j][0] > klad0[i, j][1] + 200:
                red_arrow = True
        if red_arrow:
            return region

def get_direction_pixel(region):
    all_pixels_x = []
    all_pixels_y = []
    for i,j in region.coords:
        if klad0[i, j][1] > klad0[i, j][2] + 150:
            all_pixels_x.append(i)
            all_pixels_y.append(j)
    if len(all_pixels_x) == 0:
        return None
    return sum(all_pixels_x)/len(all_pixels_x), sum(all_pixels_y)/len(all_pixels_y)
        
def get_direction_vector(region):
    pixel = get_direction_pixel(region)
    if pixel is None:
        return None
    return (pixel[0] - region.centroid[0],  pixel[1] - region.centroid[1])

def point_in_box(point, box):
    return point[0] > box[0] and point[0] < box[2] and point[1] > box[1] and point[1] < box[3]

def get_next_object(regions, region):
    if get_direction_pixel(region) is None:
        return None
    current_point = np.array(get_direction_pixel(region), dtype=float)
    direction = np.array(get_direction_vector(region))
    i = 0
    while i < 40:
        i += 1
        current_point += direction
        for next_region in regions:
            if point_in_box(current_point, next_region.bbox) and next_region != region:
                return next_region
    print("Failed to find next object")
    return None

def fill_region(region, color):
    currentAxis = plt.gca()
    currentAxis.add_patch(Rectangle((region.bbox[1], 
                                     region.bbox[0]),
                                    region.bbox[3] - region.bbox[1],
                                    region.bbox[2] - region.bbox[0],
                                    fill=False, hatch='/', edgecolor=color))

def find_final(regions):
    next_object = get_red_arrow(regions)
    current_object = next_object
    while next_object is not None:
        current_object = next_object
        fill_region(current_object, "red")
        next_object = get_next_object(regions, current_object)
        

    fill_region(current_object, "blue")
    plt.imshow(klad0)

find_final(regions)
```
